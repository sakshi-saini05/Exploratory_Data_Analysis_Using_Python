# Columns Conversion 

```python
from datetime import datetime as dt
```

```python
df["Date"]=pd.to_datetime(df["Date"])
df["Time"]=pd.to_datetime(df["Time"])
```

```python
df.dtypes
```

# Extraction from Date columns

```python
df["Month"]=df.Date.dt.month_name()  ## extracting month 
df["day"]=df.Date.dt.day   # extracting day 
df["day_of_week"]=df.Date.dt.day_name()  # extracting week of the  day 
df["hour"]=df.Time.dt.hour # extracting hour 

```

```python
df.head(2)
```

# Drop the irrevalant columns : Invoice Id and City because City and Branch telling the same thing i.e each branch indicates only one city 

```python
df.drop(columns=["Invoice ID","City"],inplace=True)
```
