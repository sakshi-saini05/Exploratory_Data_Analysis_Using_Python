# Minimization of messy Pie Chart

# Here the pie chart seems messy and the number of categories is high so we are reducing the category by fetching the top 5 deadly state and bottom 5 deadly state

# Fetch the top 5 Deadly states

```python
df_top=df.head(5)[["State/UTs","Total Cases"]]
```

# Fetch the bottom 5 least states


```python
df_bottom=df.tail(5)[["State/UTs","Total Cases"]]
```

```python
plt.figure(figsize=(10,15))
plt.subplot(1,2,1)
plt.pie(df_top["Total Cases"],labels=df_top["State/UTs"],autopct="%.f%%")
plt.title("top 5 deadly cases")
plt.subplot(1,2,2)
plt.pie(df_bottom["Total Cases"],labels=df_bottom["State/UTs"],autopct="%.f%%")
plt.title("bottom 5 least cases")
```

# top 5 active and death states

```python
df1=df.sort_values(by="Active")
df1=df.sort_values(by="Deaths")
```

```python
df1_top5=df1.head(5)[["State/UTs","Active"]]
df1_bottom5=df1.head(5)[["State/UTs","Deaths"]]
```

```python
plt.figure(figsize=(10,15))
plt.subplot(1,2,1)
plt.pie(df1_top5["Active"],labels=df1_top5["State/UTs"],autopct="%.f%%")
plt.title("top 5 active cases")
plt.subplot(1,2,2)
plt.pie(df1_bottom5["Deaths"],labels=df1_bottom5["State/UTs"],autopct="%.f%%")
plt.title("top 5 deaths cases")
```

# Recomendation to the government :Lockdown on States

- lockdown 1:complete lockdown
- lockdown 2:wear mask and only essential services are allowed to open
- lockdown 3: should wear the mask

```python
df["Lockdown_status"]=df["Total Cases"].apply(lambda x: "lockdown 3"if ((x>=7491) & (x<=317761)) else ("lockdown 2" if ((x>=317761) & (x<=996689)) else "lockdown 1"))
```

```python
df
```

# Recovered Cases

```python
df["Recovered Cases"]=df["Total Cases"]-(df["Active"]+df["Deaths"])
```

# columns Re-ordering

```python
df[['State/UTs','Total Cases', 'Active', 'Deaths','Recovered Cases','Active Ratio (%)','Death Ratio (%)', 'Lockdown_status']]
```

# Recovered ratio

```python
df["Recovered Ration(%)"]=round(df["Recovered Cases"]*100/df["Total Cases"],2)
```

```python
df
```

# As from central government point of view central govt give fund to the states acc to the number of deaths in that states

```python
plt.figure(figsize=(10,12))
d=df.sort_values(by="Deaths",ascending=False)
sns.barplot(x="State/UTs",y="Deaths",data=d)
plt.xticks(rotation=90);
```

```python
df["Death Ratio (%)"].agg(["max","min"])
```

# fund allocation to which states

```python
df["fund_allocation_status"]=df["Death Ratio (%)"].apply(lambda x:"fund allocated" if x>1.5 else "no fund")
```

```python
df
```

# Comparison Between Active and Deaths

```python
plt.figure(figsize=(15,10))
plt.plot(df["State/UTs"],df["Active"],color="green",marker="^",label="active")
plt.plot(df["State/UTs"],df["Deaths"],color="red",marker="o",label="Deaths")
plt.legend()
plt.xticks(rotation=90);
```

# Insights:
- Deaths in Maharashtra,UttarPradesh,Delhi,Chhattisgarh,Gujarat,Madhya Pradesh,Haryana,Bihar,Punjab having more number of deaths so the goverment provide more facilities of the hospital for the treatment of the patients
